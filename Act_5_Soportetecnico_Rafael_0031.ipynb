```python
# UIII Act 5
# Simulación de inteligencia artificial con if, else y elif

print("================================")
print(" IA DE SOPORTE TECNICO ")
print("================================")

# EJECUCION 1
print("\n===== EJECUCION 1 =====")

problema = input("Problema: ")
internet = input("Internet (si/no): ")
bateria = int(input("Batería: "))

if bateria < 20:
    print("Batería baja")

if internet == "si":
    print("Internet correcto")
else:
    print("Revisar módem")

if problema == "lento":
    print("Limpiar archivos")

elif problema == "virus":
    print("Instalar antivirus")

elif problema == "pantalla azul":
    print("Error del sistema")

else:
    print("Revisar cargador")

print("Carga restante:", 100 - bateria, "%")


# EJECUCION 2
print("\n===== EJECUCION 2 =====")

problema = input("Problema: ")
internet = input("Internet (si/no): ")
bateria = int(input("Batería: "))

if bateria < 20 and internet == "no":
    print("Sin internet y batería baja")

if problema == "lento":
    print("Limpiar archivos")

elif problema == "virus":
    print("Instalar antivirus")

elif problema == "pantalla azul":
    print("Error del sistema")

else:
    print("Revisar cargador")

print("Carga restante:", 100 - bateria, "%")


# EJECUCION 3
print("\n===== EJECUCION 3 =====")

problema = input("Problema: ")
bateria = int(input("Batería: "))

if bateria >= 80:
    print("Batería en buen estado")

if problema == "virus":
    print("Instalar antivirus")

else:
    print("Diagnóstico terminado")


# EJECUCION 4
print("\n===== EJECUCION 4 =====")

problema = input("Problema: ")
internet = input("Internet (si/no): ")

if problema == "no enciende" and internet == "no":
    print("Revisar cargador y módem")

else:
    print("Soporte finalizado")
```
