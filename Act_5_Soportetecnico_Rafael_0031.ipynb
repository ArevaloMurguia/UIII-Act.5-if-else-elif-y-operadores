```python
# ============================================
# SIMULADOR DE INTELIGENCIA ARTIFICIAL
# Tema: Soporte Técnico
# Rafael Arevalo NL = 6
# Fecha: 18/05/2026
# Clase Investigador No = 6
# ============================================

print("====================================")
print(" IA SOPORTE TÉCNICO ")
print("====================================")

print("\nHola técnico.")
nombre = input("Ingresa tu nombre: ")

print("\nSelecciona el área de soporte")
print("1. Computadoras")
print("2. Redes")
print("3. Software")

area = input("Opción: ")

# -------------------------
# ÁREA COMPUTADORAS
# -------------------------
if area == "1":
    print("\nÁrea seleccionada: Computadoras")

    bateria = int(input("Nivel de batería (%): "))
    temperatura = int(input("Temperatura del equipo: "))

    if bateria >= 50 and temperatura <= 60:
        print("\nIA:", nombre + ", el equipo funciona correctamente.")
        print("No se detectan fallas.")

    elif bateria >= 20 and temperatura <= 75:
        print("\nIA: Rendimiento aceptable.")
        print("Se recomienda revisión preventiva.")

    else:
        print("\nIA: Alerta crítica.")
        print("El equipo necesita mantenimiento urgente.")

# -------------------------
# ÁREA REDES
# -------------------------
elif area == "2":
    print("\nÁrea seleccionada: Redes")

    señal = int(input("Nivel de señal (%): "))
    errores = int(input("Cantidad de errores detectados: "))

    if señal >= 80 and errores == 0:
        print("\nIA: Red funcionando perfectamente.")

    elif señal >= 50 and errores <= 3:
        print("\nIA: Red estable con pequeñas fallas.")

    else:
        print("\nIA: Conexión inestable.")
        print("Se requiere soporte inmediato.")

# -------------------------
# ÁREA SOFTWARE
# -------------------------
elif area == "3":
    print("\nÁrea seleccionada: Software")

    memoria = int(input("Uso de memoria (%): "))
    actualizado = input("¿El sistema está actualizado? (si/no): ")

    if memoria <= 70 and actualizado == "si":
        print("\nIA: Software optimizado.")
        print("El sistema opera correctamente.")

    elif memoria <= 90 or actualizado == "si":
        print("\nIA: Software funcional con riesgo moderado.")

    else:
        print("\nIA: Error crítico.")
        print("Se necesita actualización inmediata.")

# -------------------------
# OPCIÓN INVÁLIDA
# -------------------------
else:
    print("\nIA: Opción no válida.")
    print("Reinicia el programa.")
```
